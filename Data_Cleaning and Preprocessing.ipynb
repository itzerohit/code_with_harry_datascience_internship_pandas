# Handling Missing Values

. Check for Missing Values

```python
df.isnull() # True for NaNs
```

```python
df.isnull().sum() # count missing per column 
```

# Drop Missing Data

```python
df.dropna()  # Drop rows with *any* missing values
```

```python
df.dropna(axis=1) # Drop columns with missing values
```

```python
df.fillna(10) # Jaha jaha bhi nan value hoga waha pe 10 placed ho jayega 
```

```python
df["Age"].fillna(df["Age"].mean()) # Replacing the nan value data with the mean
```

```python
df
```

```python
df.fillna(method = "ffill")     # This is basically a forward fill
```

```python
df.fillna(method = "bfill") # This is basically a backward fill
```

```python
df.ffill() # THis is the current version to fill forward number 
```

```python
df.bfill()  # This is the current version to fill by using backward number
```

# String Operations with .str

Works like vectorized string methods and returns a pandas series 

```python
# Convert all names to lowercase.
df["Name"].str.lower()
```

```python
df["Name"].str.upper()
```

```python
# Check if 'delhi' is in the city name, case-insensitive
```

```python
df["City"].str.contains("delhi", case=False)
```

```python
# Outputs a pandas series where each element is a list
df["Email"].str.split("@")
```

This is where a Python list comes into play, but the outer object is still a pandas series


# Type Conversions with .astype()

# Applying Functions

.apply() -- Apply any function to rows or column

```python
df2
```

```python
df2["Age Group"] = df2["Age"].apply(lambda x: "Adult" if x > 25 else "Minor")
```

```python
df2
```

.map() -- Element-wise mapping for Series

```python
gender_map = {"M": "Male", "F": "Female", "O": "Others"}
df2["Gender"] = df2["Gender"].map(gender_map)
```

```python
df2
```

```python
df
```

.replace() -- Replace specific value

```python
df["City"] = df["City"].replace({"Delhi":"DELHI"})
```

```python
df
```
